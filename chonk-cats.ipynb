```python
# Import libraries

import numpy as np
import tensorflow as tf
import pandas as pd
import sklearn

from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# Prepare data

data = pd.read_csv('C:/Users/Student/Desktop/IzzyL/Projects/ChonkCats/cats_dataset-og.csv')

one_hot = pd.get_dummies(data["Label"])
x_cat = data.drop("Label", axis=1)
data = pd.concat([data, one_hot], axis=1)
y_chonk = one_hot

x_cat = pd.get_dummies(x_cat)

# Train test split

x_cat_train, x_cat_test, y_chonk_train, y_chonk_test = train_test_split(x_cat, y_chonk, test_size=0.30, shuffle=True)

x_cat_train = x_cat_train.astype(np.float32).to_numpy()
x_cat_test = x_cat_test.astype(np.float32).to_numpy()
y_chonk_train = y_chonk_train.astype(np.float32).to_numpy()
y_chonk_test = y_chonk_test.astype(np.float32).to_numpy()

# Model

model = Sequential()

model.add(Flatten())
model.add(Dense(units=4000, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=2000, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=500, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=250, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=7, activation='softmax'))

# Compile and fit model

print(x_cat_train.shape)
print(y_chonk_train.shape)

model.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), metrics=['accuracy'])

model.fit(
    x=x_cat_train,
    y=y_chonk_train,
    batch_size=20,
    epochs=50,
    validation_data=(x_cat_test, y_chonk_test),
)

# Evaluate model

scores = model.evaluate(x_cat_test, y_chonk_test, verbose=1)
print("Test accuracy: ", scores[1])

# Save model

model.save('chonk-chart-model.h5')
```
